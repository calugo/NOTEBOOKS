```python
def plotswav(sig):
    import matplotlib.pyplot as plt
    import numpy as np
    import wave
    import sys

    #lswav=glob.glob("recordings/*wav")
    #print lswav
    spf = wave.open(sig,'r')

#Extract Raw Audio from Wav File
    signal = spf.readframes(-1)
    signal = np.fromstring(signal, 'Int16')


    #If Stereo
    if spf.getnchannels() == 2:
        print 'Just mono files'
        sys.exit(0)

    plt.figure(1)
    plt.title('Signal Wave...')
    plt.plot(signal)
#plt.show()
```

```python
def plotswavdet(sig,n1,n2):
    import matplotlib.pyplot as plt
    import numpy as np
    import wave
    import sys

    #lswav=glob.glob("recordings/*wav")
    #print lswav
    spf = wave.open(sig,'r')

#Extract Raw Audio from Wav File
    signal = spf.readframes(-1)
    signal = np.fromstring(signal, 'Int16')


    #If Stereo
    if spf.getnchannels() == 2:
        print 'Just mono files'
        sys.exit(0)

    plt.figure(1)
    plt.title('Signal Wave...')
    plt.plot(signal[n1:n2])
#plt.show()
```
