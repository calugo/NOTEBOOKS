```python
def movingaverage(signal,window):
    import numpy as np
    values=[]
    for i in signal:
        values.append(float(i))
    
    weigths = np.repeat(1.0, window)/window
    smas = np.convolve(np.asarray(values), weigths, 'valid')
    return smas # as a numpy array
```

```python
def plotswavdet(sig,n1,n2,df,MN):
    import matplotlib.pyplot as plt
    import numpy as np
    import wave
    import sys
    from numpy.fft import fft
    from numpy.fft import ifft

    #lswav=glob.glob("recordings/*wav")
    #print lswav
    spf = wave.open(sig,'r')

#Extract Raw Audio from Wav File
    signal = spf.readframes(-1)
    signal = np.fromstring(signal, 'Int16')

    fr=spf.getframerate()
    print fr
    #If Stereo
    if spf.getnchannels() == 2:
        print 'Just mono files'
        sys.exit(0)
    t=[(1.0/float(fr))*tj for tj in xrange(len(signal[n1:n2]))]
    plt.figure(1)
    plt.title('Signal Wave...')
    plt.plot(t,signal[n1:n2])
    
    
    fs=fft(np.asarray(signal[n1:n2]))
    
    fr=[(float(fr)/len(signal[n1:n2]))*tj for tj in xrange(len(signal[n1:n2]))]
    
    fig3, ax3 = plt.subplots()
    rects3 = ax3.plot(np.log(fr),np.log(abs(fs)/len(signal[n1:n2])),'-' , color='b')
    
    filta=[0.0*qjj for qjj in xrange(len(signal[n1:n2]))]
    for j in range(0, int( 0.5*(len(signal[n1:n2])-1) ) ):
        if j<((0.5)*len(signal[n1:n2])-df) and (j>MN):
            filta[j]=1.0
            filta[len(signal[n1:n2])-1-j]=1.0   
    
    fig4, ax4 = plt.subplots()
            
    rects4 = ax4.plot(fr,filta,'-' , color='b')
    
    yf=[i*j for i,j, in zip(fs,filta)]
    #print len(yf), len(fs)
    
    fig5, ax5 = plt.subplots()
    rects5 = ax5.plot(np.log(fr),np.log(yf),'-' , color='b')
    
    
     
    fig6, ax6 = plt.subplots()
    rects6 = ax6.plot(t,ifft(yf),'-' , color='b')
#plt.show()
```
